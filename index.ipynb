```python
import pandas as pd
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
receitas = pd.read_csv('receita/receita_geral_RS_total.csv')
despesas = pd.read_csv('despesa/despesa_geral_RS_total.csv')

receita_ano = pd.read_csv('receita/receita_geral_RS.csv')
despesa_ano = pd.read_csv('despesa/despesa_geral_RS.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=receitas.head(10), x='PARTIDO', y='RECEITA')
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x / 1e6:.1f}M'))
plt.title('Receita Total por Partido (2016-2024)')
plt.ylabel('Total Receita (R$)')
plt.show()
```

```python
partidos = receitas.head(5)['PARTIDO'].unique()
filter = receita_ano[receita_ano['PARTIDO'].isin(partidos)]

plt.figure(figsize=(10, 6))  # Aumenta o tamanho da figura para melhorar o layout
sns.lineplot(data=filter, x='ANO', y='RECEITA', hue='PARTIDO', marker='o')

# Ttulo e rtulos
plt.title('Receita por Ano e Partido')
plt.ylabel('Receita (R$)')
plt.xlabel('Ano')

# Definir os anos corretamente no eixo X e girar rtulos
anos = list(range(2017, 2025))  # Anos de 2016 a 2024
plt.xticks(ticks=anos, rotation=45)  # Gira os rtulos para evitar sobreposio
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x / 1e6:.1f}M'))


# Adiciona espaamento para evitar corte de rtulos
plt.tight_layout()

# Exibe o grfico
plt.show()
```
