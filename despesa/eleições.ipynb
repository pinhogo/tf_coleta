```python
import pandas as pd
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
tables = ['despesa_anual_2017_RS.csv', 'despesa_anual_2018_RS.csv', 'despesa_anual_2019_RS.csv', 'despesa_anual_2020_RS.csv', 'despesa_anual_2021_RS.csv', 'despesa_anual_2022_RS.csv', 'despesa_anual_2023_RS.csv', 'despesa_anual_2024_RS.csv']
```

```python
dataframes = []

for file in tables:
    df = pd.read_csv(file, encoding='latin1', sep=';')
    dataframes.append(df)

data = pd.concat(dataframes, ignore_index=True)

data.head()
```

```python
data.drop(['TP_DESPESA','SG_UF','NR_ZONA','DS_GASTO','DT_PAGAMENTO','DS_FONTE_DESPESA','SQ_DESPESA', 'HH_GERACAO','AA_AIDF','VR_GASTO', 'VR_DOCUMENTO', 'NR_AIDF','NM_PARTIDO', 'DS_TP_DOCUMENTO','NR_DOCUMENTO','DS_TP_FORNECEDOR','CD_TP_ESFERA_PARTIDARIA','CD_TP_DOCUMENTO', 'NR_CPF_CNPJ_FORNECEDOR', 'NR_CNPJ_PRESTADOR_CONTA', 'DT_GERACAO', 'DS_TP_ESFERA_PARTIDARIA'], axis=1, inplace=True)
```

```python
data.sample(n=5)
```

```python
data = data[data['VR_PAGAMENTO'] != '0']
```

```python
data.sample(n=5)
```

```python
data.replace('#NULO#', np.nan, inplace=True)
```

```python
sns.heatmap(data.isna())
```

```python
num_nulos_nm_municipio = data['CD_TP_FORNECEDOR'].isna().sum()
print(num_nulos_nm_municipio)
```

```python
partidos_com_nulos = data.groupby('SG_PARTIDO')['NM_FORNECEDOR'].apply(lambda x: x.isna().sum())
partidos_com_nulos = partidos_com_nulos[partidos_com_nulos == 0]

print(partidos_com_nulos)
```

```python
def format_value(value):
    if isinstance(value, str) and value.startswith(','):
        return '0' + value
    return value
data = data.applymap(format_value)

data['VR_PAGAMENTO'] = data['VR_PAGAMENTO'].str.replace(',', '.').astype(float)
```

```python
total_vr_pagamento_por_partido = data.groupby('SG_PARTIDO')['VR_PAGAMENTO'].sum()
total_vr_pagamento_por_partido = total_vr_pagamento_por_partido.sort_values(ascending=False)

print(total_vr_pagamento_por_partido)   
```
